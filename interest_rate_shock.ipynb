```python
import matplotlib.pyplot as plt
from rank import RANK
from tank import TANK
import numpy as np
import params

np.set_printoptions(suppress=True, precision=10, linewidth=250)
plt.style.use('default')
plt.rcParams.update({
    "font.family": "serif",  # use serif/main font for text elements
    "text.usetex": True,     # use inline math for ticks
    "pgf.rcfonts": False     # don't setup fonts from rc parameters
})
```

```python
# paramaters
PAGE_WIDTH = 6.5 # 8.5 in page - 2*1 in margins
T = 25 # max time
i = 1 # index of shock we observe

save = True
```

```python
# initialize models
rank = RANK(params)
tank = TANK(params)

# irfs
rank_irf = rank.run_impulse_response(T, i)
tank_irf = tank.run_impulse_response(T, i)
```

```python
# pi response
fig, ax = plt.subplots()
fig.set_figwidth(PAGE_WIDTH)
fig.set_figheight(PAGE_WIDTH / 2.5)
ax.set_xlabel(r'$t$')
ax.set_ylabel(r'\% Deviation from Steady State')

ax.plot(rank_irf.T[:, 6], label=RANK.LABELS[6])
ax.plot(tank_irf.T[:, 12], label=TANK.LABELS[12])

ax.legend()
fig.tight_layout(pad=0)
if save: fig.savefig('../document/graphs/i_shock_pi.pdf', bbox_inches='tight', pad_inches=0.05)
```

```python
# r response
fig, ax = plt.subplots()
fig.set_figwidth(PAGE_WIDTH)
fig.set_figheight(PAGE_WIDTH / 2.5)
ax.set_xlabel(r'$t$')
ax.set_ylabel(r'\% Deviation from Steady State')

ax.plot(rank_irf.T[:, 10], label=RANK.LABELS[10])
ax.plot(tank_irf.T[:, 16], label=TANK.LABELS[16])

ax.legend()
fig.tight_layout(pad=0)
if save: fig.savefig('../document/graphs/i_shock_r.pdf', bbox_inches='tight', pad_inches=0.05)
```
